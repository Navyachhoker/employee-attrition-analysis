finding correlation

```python
# Convert target column to numeric
df['Attrition_numerical'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Select only numerical columns
numerical = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation
corr = numerical.corr()

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=False, cmap='viridis')

plt.title('Correlation Heatmap')
plt.show()
```

```python
attrition = df.drop(['Attrition_numerical'], axis=1)
```

```python
categorical = []
for col, value in attrition.items():
    if value.dtype == 'object':
        categorical.append(col)
```

```python
attrition_cat = df[categorical]
attrition_cat = attrition_cat.drop('Attrition', axis=1)
attrition_cat = pd.get_dummies(attrition_cat)

numerical = df.select_dtypes(exclude='object').columns
attrition_num = df[numerical]

attrition_final = pd.concat([attrition_num, attrition_cat], axis=1)
```

```python
attrition_final.head()
```

splitting data 

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
target =df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    attrition_final, target,
    test_size=0.2,
    random_state=0,
    stratify=target 
)
```

training the data

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Random Forest model
rf = RandomForestClassifier(
    n_estimators=1000,
    max_depth=4,
    min_samples_leaf=2,
    max_features='sqrt',
    random_state=0,
    n_jobs=-1
)

# Train model on original data (no SMOTE)
rf.fit(X_train, y_train)

# Predict
y_pred = rf.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("="*50)
print(classification_report(y_test, y_pred))
```

visaulizing feature importance in Random forest

```python


importance_df = pd.DataFrame({
    'Feature': attrition_final.columns,
    'Importance': rf.feature_importances_
})

importance_df = importance_df.sort_values(by='Importance', ascending=False).head(15)

trace = go.Bar(
    x=importance_df['Importance'],
    y=importance_df['Feature'],
    orientation='h',
    marker=dict(
        color=importance_df['Importance'],
        colorscale='Viridis',
        showscale=True
    )
)

layout = go.Layout(
    title='Top Feature Importance',
    height=700,
    margin=dict(l=200)
)

fig = go.Figure(data=[trace], layout=layout)
py.iplot(fig)
```
